# What is Machine Learning?

- *Machine learning is the field of study that gives computers the ability to learn without being explicitly programmed. (Arthur Samuel, 1959)*

- *A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with experience E. (Tom Mitchell, 1997)*

# Types of Machine Learning Systems

ML Systems are categorized based on:
* Learning Mode (how they are guided during training) => Supervised learning, Unsupervised learning, Semi-supervised learning, Self-supervised learning, Reinforcement learning
* Training Mode => Ability for learning incrementally => Batch vs Online learning
* Modeling Mode => By comparing new data point with known data points or by building a predictive model based on detected patterns in training data) => Instance-based vs Model-based learning

### Semi-supervised learning

* In Semi-supervised learning, training data is partially labeled.
* Combines Supervised and Unsupervised algorithms
* Eg: Google Photos (Clustering to identify photos with same person, unlabeled persons labeled with most common label in its cluster)

### Self-supervised learning

* Self-supervised learning involves
    * generation of a fully labeled dataset from a fully unlabeled dataset
    * training of a model on the above dataset
    * tweaking of the model to serve the particular ML task
    * use of a supervised learning algorithm for fine-tuning in the final step
* Eg: Pet Image Classification task
    * Train Pet Image Repairing Model with
        * inputs to the model => Unlabeled pet images with mask
        * labels => Original pet images
    * Once it is performing well (repairs pet images correctly), tweak model architecture to predit pet species instead of repairing images
    * Fine-tune model on labeled dataset (inputs => Original pet images, labels => pet species)
* Self-supervised learning tasks include Classification, Regression etc.

### Reinforcement learning

* In Reinforcement learning, the learning system (agent)
    * observes environment & perform actions based on a policy
    * get rewards or penalties in return
    * updates policy based on rewards or penalties
    * iterates till an optimum policy for a given task
* Eg: DeepMind's AlphaGo

## Training Mode (Batch vs Online learning)

### Batch learning

* Model trained using all available data, typically offline => Offline learning
* Deployed in production without learning anymore from new data
* Data drift or Model decay => Needs to retrain regularly on up-to-date data (train from scratch on old data + new data)
* Training on full dataset is slow & resource-heavy => costly & slow adaptation to change
* Eg: Random Forests => can only be trained from scratch on the full dataset

### Online learning

* Model trained incrementally on new data using mini-batches online => Online learning
* Each learning step is fast & resource-light => cheap & fast adaptation to change
* Eg: Gradient descent algorithm

## Modeling Mode (Instance-based vs Model-based Learning)

### Instance-based learning

* Prediction based on a *measure of similarity* with known instances
* Need to deploy a copy of training set to production => best for small dataset models; not recommended for models with big datasets

```python
from IPython.display import Image, display, Markdown
display(Image(filename="data/images/machine_learning_landscape_16.png", width=500))
display(Markdown("*Instance-based learning: in this example we consider the class of the three nearest neighbors in the training set*"))
```

### Model-based learning

* Instead of predicting from a set of instances, a suitable model is selected to capture the relationships in the training data
* After fitting this model on training data, it is used to make predictions on new data

```python
from IPython.display import Image, display, Markdown
display(Image(filename="data/images/machine_learning_landscape_17.png", width=500))
display(Markdown("*Model-based learning*"))
```
