# What is Machine Learning?

- *Machine learning is the field of study that gives computers the ability to learn without being explicitly programmed. (Arthur Samuel, 1959)*

- *A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with experience E. (Tom Mitchell, 1997)*

# Examples of Applications

| Application | Task | Method |
| :---------- | :--- | :----- |
| *Analyzing images of products on a production line* | Image Classification | CNNs (ResNet, EfficientNet), Vision Transformers |
| *Detecting tumors in brain scans* | Semantic Image Segmentation | CNNs (U-Net, Mask R-CNN), Vision Transformers |
| *Automatically classifying news articles* | Text Classification (NLP) | Transformers (BERT, RoBERTa), RNNs (legacy) |
| *Summarizing long documents automatically* | Text Summarization (NLP) | Transformers (T5, BART), RNNs (legacy) |
| *Estimating genetic risk from long DNA sequences* | Long-sequence pattern discovery | State Space Models (SSMs), Transformers |
| *Forecasting company revenue* | Regression / Time Series Forecasting | Linear Models, SVMs, Random Forests, Gradient Boosting, ANNs |
| *Making an app react to voice commands* | Speech Recognition | CNNs (spectrograms), RNNs, Transformers |
| *Detecting credit card fraud* | Anomaly Detection | Isolation Forests, GMMs, Autoencoders, One-Class SVM |
| *Segmenting clients by purchasing behavior* | Clustering | K-means, DBSCAN, GMM, Hierarchical Clustering |
| *Visualizing high-dimensional data* | Dimensionality Reduction | PCA, t-SNE, UMAP, Autoencoders |
| *Recommending products to users* | Recommender Systems | Collaborative Filtering, Matrix Factorization, Neural Recommenders |
| *Creating a chatbot or personal assistant* | Question Answering / Dialogue (NLP) | Transformers (LLMs), RAG, Agent-based systems |
| *Building an intelligent game bot* | Reinforcement Learning | Q-Learning, DQN, Decision Transformers |

# Types of Machine Learning Systems

ML Systems are categorized based on:
* Learning Mode (how they are guided during training) => Supervised learning, Unsupervised learning, Semi-supervised learning, Self-supervised learning, Reinforcement learning
* Training Mode => Ability for learning incrementally => Online vs Batch learning
* Modeling Mode => By comparing new data point with known data points or by building a predictive model based on detected patterns in training data) => Instance-based vs Model-based learning
